1. Propositional Logic Operations:

The AND operation returns True if both operands are True; otherwise, it returns False.

The OR operation returns True if at least one of the operands is True; it returns False only if both are False.

The NOT operation returns True if the operand is False, and it returns False if the operand is True

The IMPLIES operation returns False only if the first operand is True and the second operand is False. In all other cases, it returns True.

```python
def and_operation(p, q):
    return p and q

def or_operation(p, q):
    return p or q

def not_operation(p):
    return not p

def implies_operation(p, q):
    return not p or q
```

2. Logical Statements

```python
def evaluate(statement, values):
    """Evaluate a logical statement based on given truth values."""
    # Replace variables in the statement with their truth values
    for var, val in values.items():
        statement = statement.replace(var, str(val))

    # Evaluate the statement using eval (ensure the statement is safe)
    return eval(statement)
```

```python
print(and_operation(True, False))
print(or_operation(True, False))
print(not_operation(True))
print(implies_operation(True, False))
```

3. Functions for Quantifiers

```python
def forall(predicate, domain):
    """Evaluate a predicate for all elements in the domain."""
    return all(predicate(x) for x in domain)

def exists(predicate, domain):
    """Evaluate if there exists at least one element in the domain that satisfies the predicate."""
    return any(predicate(x) for x in domain)
```

```python
# Example predicate
def is_even(x):
    return x % 2 == 0

# Domain of numbers
domain = range(1, 11)

# Check for all even numbers
print("All numbers are even:", forall(is_even, domain))

# Check if there exists an even number
print("Exists an even number:", exists(is_even, domain))
```

```python
class SimpleAIAgent:
    def __init__(self):
        pass

    def decide(self, condition):
        """Decide action based on a logical condition."""
        if condition:
            return "Action A: Proceed."
        else:
            return "Action B: Halt."

# Example scenario
agent = SimpleAIAgent()
condition = and_operation(True, False)  # Example condition
decision = agent.decide(condition)
print("AI Decision:", decision)
```
