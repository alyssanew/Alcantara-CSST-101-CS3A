# **Visualize a Probability Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10  # number of trials (e.g., tests)
p = 0.3  # probability of success (e.g., probability of disease given symptoms)

x = np.arange(0, n + 1)


pmf = binom.pmf(x, n, p)


plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='skyblue', alpha=0.7)
plt.title('Binomial Distribution PMF', fontsize=16)
plt.xlabel('Number of Successes', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.xticks(x)
plt.grid(axis='y')

plt.show()
```
